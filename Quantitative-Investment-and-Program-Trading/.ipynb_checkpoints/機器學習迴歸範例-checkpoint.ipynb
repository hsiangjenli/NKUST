# 以下處理技術指標

需要10個特徵 ('MA10' , 'WMA10', 'MOM' , 'K' , 'D', 'RSI-10' , 'dea值' , '威廉指標R' , 'AD線' , 'CCI指標' )

```python
# 開高低收量
# 各用一個變數儲存
High  = df["High"]   # 最高價
Low   = df["Low"]   # 最低價
Open  = df["Open"]   # 開盤價
Close = df["Close"]   # 收盤價
Volume= df["Volume"] # 成交量
```

##### 技術指標--均線

```python
df['MA10']  = talib.MA(Close, timeperiod=10)
df['WMA10'] = talib.WMA(Close, timeperiod=10)
df['EMA10'] = talib.EMA(Close, timeperiod=10)
```

##### 技術指標--動量線

```python
df['MOM'] = talib.MOM(Close, timeperiod=10)
```

##### 技術指標--KD

```python
fastk, fastd = talib.STOCHF( High ,Low, Close, fastk_period=5, fastd_period=3, fastd_matype=0)
df['K'] = fastk
df['D'] = fastd
```

##### 技術指標--RSI

```python
df['RSI-5']  = talib.RSI(Close,timeperiod=5)
df['RSI-10'] = talib.RSI(Close,timeperiod=10)
```

##### 技術指標--MACD

```python
dif, macd, dea = talib.MACD(Close,12,26,9)
df['dif值']= dif
df['macd值']= macd
df['dea值']= dea
```

##### 技術指標--威廉R指標

```python
df['威廉指標R'] = talib.WILLR(High, Low, Close, timeperiod=14)
```

##### 技術指標--AD線

```python
df['AD線'] = talib.AD(High, Low, Close, Volume)
```

##### 技術指標--CCI線

```python
df['CCI指標'] = talib.CCI(High, Low, Close, timeperiod=14)
```

##### 計算報酬率  (LABEL採用)

```python
R = (Close - Close.shift(1))/Close.shift(1)
# R = (Close.shift(-1)-Close)/Close # (今天-昨天)/昨天，一樣的結果

df['報酬率(Return)']= R.shift(-1) # 位移1日
df['報酬率(Return)']
```

# 隨機森林模型

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestRegressor

# 建立RandomForestRegressor模型
randomForestModel = RandomForestRegressor(n_estimators=100, criterion = 'mse')
# 使用訓練資料訓練模型
randomForestModel.fit(X_train, y_train)
# 使用訓練資料預測
predicted=randomForestModel.predict(X_test)
```

```python
from sklearn import metrics
# print('R2 score: ', randomForestModel.score(X_test, y_test))
mse = metrics.mean_squared_error(y_test, predicted)
print('MSE score: ', mse)
```

```python
years = y_test.index
pops =y_test
deaths = predicted

lines = plt.plot(years,pops,years,deaths)
plt.xlabel('x')
plt.ylabel('y')
```

## LSTM

```python
# LSTM
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout,BatchNormalization
from keras.models import Sequential
times= 50
unit= 100
my_model = Sequential()
my_model.add(LSTM(units = unit, input_shape= (10,1)))
my_model.add(Dropout(0.5))
my_model.add(Dense(100, activation='relu', input_dim= (10,1)))
my_model.add(Dense(1))
my_model.compile(loss = 'mse',optimizer = 'adam', metrics = ['mean_squared_error'])
my_model.summary()
history= my_model.fit(X_train,y_train,epochs= times)
submission_pfs = my_model.predict(X_test)
submission_pff = my_model.predict(y_test)

#print('The next 0.2 real price is ',submission_pfs)
print('The future prices of next year is', submission_pff)
# ----------------------------------------------------------------------------
```
